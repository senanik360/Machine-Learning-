### Importing the Libraries

```python
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd
```

### importing the dataset

```python
df = pd.read_csv('online.csv')
```

```python
df = pd.DataFrame(df)
df.head()
```

```python
df.shape
```

```python
df.isnull().sum() #checking null values
```

```python
#  if missing anything or null
# missing = df.Administration.mean()
# df.Adminstration = df.Adminstration.fillna(missing)
```

## Separe x,y

```python
x = df.drop(['Profit'], axis=1)
```

```python
y = df['Profit']
```

## One hot Encoding

```python
# conver the column into categorical columns
city = pd.get_dummies(x['Area'], drop_first=True)
city
```

```python
# Drop the Area column
x = x.drop('Area', axis=1)
x
```

```python
# concatenation 
x = pd.concat([x,city], axis=1)
```

```python
x
```

# Splitting the dataset into the Training set and Test Set

### import library 


```python
from sklearn.model_selection import train_test_split 
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x,y,test_size = 0.25, random_state = 0)
```

## Fitting Multiple Linear Regression to the Training Set

```python
from sklearn.linear_model import LinearRegression
```

```python
regressor = LinearRegression()
```

```python
regressor.fit(xtrain, ytrain)
```

```python
xtest
```

```python
ytest
```

## Predicting the test set result

```python
pred = regressor.predict(xtest)
```

```python
pred
```

```python
regressor.score(xtest,ytest)
```

# R-Squared Value

```python
from sklearn.metrics import r2_score
```

```python
score = r2_score(ytest, pred)
score
```
